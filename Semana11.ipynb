Iniciando com o Pandas

```python
import pandas as pd
import numpy as np
```

```python
lista_alunos = ["Fulano","Ciclano","aluno1", "aluno 2", "aluno 3"]
```

```python
nome = pd.Series(lista_alunos)
```

Criando um DataFrame

```python
nome
```

```python
alunos = pd.DataFrame(nome, columns=["Nome"])
```

```python
alunos
```

```python
alunos['AV1'] = [4.0, 5.0, 6.0, 7.0, 1.0]
```

```python
alunos
```

```python
alunos['AV2'] = [9.0, 3.0, 10.0, 2.0, 5.0 ]
```

```python
alunos
```

```python
alunos['Media'] = (alunos['AV1'] + alunos['AV2']) /2
```

```python
alunos
```

```python
alunos['Resultado'] = np.where(alunos['Media']>=6.0,"Aprovado", "Foi para a AV3")
```

```python
alunos
```

```python
alunos.sort_values(by='Resultado')
```

```python
alunos[alunos['Resultado'] == "Aprovado"]
```

```python
alunos.describe()
```
